```python
import numpy as np
```

```python
# Write a NumPy program to find the sum of all elements in an array [10, 20, 30]

arr = np.array([10, 20, 30])
total_sum = np.sum(arr)

print(f"Array: {arr}")
print(f"Sum of elements: {total_sum}")
```

```python
# Write a NumPy program to find the shape of the array
# [[1, 2, 3], [4, 5, 6]].

arr = np.array([[1, 2, 3], [4, 5, 6]])
print(f"Array shape: {arr.shape}")
```

```python
# Write a NumPy program to find the maximum element in [5, 12, 3, 9].

arr = np.array([5, 12, 3, 9, 20, 6, 25])
max_val = np.max(arr)

print(f"Maximum element: {max_val}")
```

```python
#Write a NumPy program to add two arrays
#[1, 2, 3] and [4, 5, 6].

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = arr1 + arr2

print(f"Array 1: {arr1}")
print(f"Array 2: {arr2}")
print(f"Resulting array: {result}")
```

```python
# Write a NumPy program to check the data type of an array [1, 2, 3].

arr = np.array([1, 2, 3])
print(f"Data type: {arr.dtype}")
```
