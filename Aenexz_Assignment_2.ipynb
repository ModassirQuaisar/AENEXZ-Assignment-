```python
def get_area(radius):
    return 3.14 * radius * radius

user_radius = float(input("Enter the radius of the circle: "))

area = get_area(user_radius)
print("The area of the circle is:", area)
```

```python
# 2. Write a function that accepts three numbers and returns the largest one.

def find_largest(num1, num2, num3):
    return max(num1, num2, num3)

print("Let's find the largest of three numbers!")
first = int(input("Enter the first number: "))
second = int(input("Enter the second number: "))
third = int(input("Enter the third number: "))

largest_num = find_largest(first, second, third)
print("The largest number is:", largest_num)
```

```python
# 3. Write a function that takes a string and returns its length without using built-in length functions.

def calculate_length(text):
    count = 0
    for char in text:
        count = count + 1  
    return count

user_input = input("Enter a word or sentence: ")

string_length = calculate_length(user_input)
print("The length of your text is:", string_length)
```

```python
#4. Write a function that checks whether a number is prime.

def is_prime(number):
    if number <= 1:
        return False
        
    for i in range(2, number):
        if number % i == 0:
            return False
            
    return True

user_num = int(input("Enter a number to check: "))

if is_prime(user_num):
    print(user_num, "is a Prime number.")
else:
    print(user_num, "is NOT a Prime number.")
```

```python
#7. Write a program to find the sum of all numbers from 1 to n using a for loop. 

n = int(input("Enter a number (n): "))

total_sum = 0

for i in range(1, n + 1):
    total_sum = total_sum + i

print("The sum of numbers from 1 to", n, "is:", total_sum)
```

```python
#8. Write a program to count how many even numbers are present in a list.

numbers = [12, 7, 9, 24, 18, 5, 2, 10]

even_count = 0

for num in numbers:
    
    if num % 2 == 0:
        even_count = even_count + 1  

print("The list of numbers is:", numbers)
print("Total even numbers found:", even_count)
```

```python
#9. Write a program to print all prime numbers between 1 and 100.

print("Prime numbers between 1 and 100:")

for num in range(2, 101):
    
    is_prime = True
    
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break          
    if is_prime:
        print(num, end=" ")
```
