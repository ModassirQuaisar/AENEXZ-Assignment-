```python
#1. Write a program to find the largest of three numbers.

def find_largest(num1, num2, num3):
    """
    Finds the largest of three numbers using conditional statements.
    """
    if (num1 >= num2) and (num1 >= num3):
        largest = num1
    elif (num2 >= num1) and (num2 >= num3):
        largest = num2
    else:
        largest = num3

    return largest
    
a = 15
b = 42
c = 8

largest_num = find_largest(a, b, c)
print(f"The largest number among {a}, {b}, and {c} is: {largest_num}")
```

```python
# 2. Write a program to check whether a given year is a leap year.

year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year.")
else:
    print(year, "is NOT a Leap Year.")
```

```python
# 3. Write a program to check if a character is a vowel or consonant. 

char = input("Enter a letter: ")

# Check if the letter is in the string of both lowercase and uppercase vowels
if char in 'aeiouAEIOU':
    print(char, "is a Vowel.")
else:
    print(char, "is a Consonant.")
```

```python
# 4. Write a program to check whether a number is divisible by both 5 and 11. 

number = int(input("Enter a number: "))

if number % 5 == 0 and number % 11 == 0:
    print(number, "is divisible by both 5 and 11.")
else:
    print(number, "is NOT divisible by both 5 and 11.")
```

```python
#6 Write a program to print the multiplication table of a given number using a while loop. 

number = int(input("Enter a number for its multiplication table: "))

multiplier = 1

print(f"\nMultiplication Table for {number}:")
while multiplier <= 10:
    result = number * multiplier
    print(number, "x", multiplier, "=", result)

    multiplier = multiplier + 1
```

```python
#7. Print a pyramid pattern: 

'''* 
  *** 
 ***** 
*******'''

rows = int(input("Enter the number of rows for the pyramid: "))

for i in range(1, rows + 1):
    spaces = " " * (rows - i)
    stars = "*" * ((2 * i) - 1)
    print(spaces + stars)
```
